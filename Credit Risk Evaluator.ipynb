```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
```

```python
# Import the data
```

```python
df = pd.read_csv('Resources/lending_data.csv')
df.head()
```

```python
# Split the data into X_train, X_test, y_train, y_test
```

```python
 # Define the X (features) and y (target) sets
X = df.drop("loan_status", axis=1)
y = df["loan_status"]
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
X_train.head()
```

```python
# Train a Logistic Regression model print the model score
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs')
classifier.fit(X_train, y_train)
classifier.score(X_test, y_test)
print(f"Logistic Regression Model Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Logistic Regression Model Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(random_state=1)
classifier.fit(X_train, y_train)
classifier.score(X_test, y_test)
print(f'Random Forest Classifier Model Training Score: {classifier.score(X_train, y_train)}')
print(f'Random Forest Classifier Model Testing Score: {classifier.score(X_test, y_test)}')
```
